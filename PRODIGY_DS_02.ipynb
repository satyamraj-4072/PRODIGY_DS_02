```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/Titanic-Dataset.csv')

# Display basic information about the dataset
print(data.info())

# Display the first few rows of the dataset
print(data.head())
```

```python
# Handling missing values
data.isnull().sum()

# Fill missing Age values with the median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing Embarked values with the mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Drop Cabin column due to high number of missing values
data.drop(columns=['Cabin'], inplace=True)

# Verify missing values have been handled
print(data.isnull().sum())
```

```python
# Distribution of survivors
sns.countplot(x='Survived', data=data)
plt.title('Distribution of Survivors')
plt.show()
```

```python
# Age distribution
plt.hist(data['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Class distribution
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Relationship between class and survival
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Gender distribution
sns.countplot(x='Sex', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
# Relationship between gender and survival
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Survival by Gender')
plt.show()
```

```python
# Sibling/Spouse and Parent/Child distribution
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='SibSp', data=data)
plt.title('Siblings/Spouses Distribution')

plt.subplot(1, 2, 2)
sns.countplot(x='Parch', data=data)
plt.title('Parents/Children Distribution')
plt.tight_layout()
plt.show()
```

```python
# Fare distribution
plt.hist(data['Fare'], bins=20)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between fare and survival
sns.boxplot(x='Survived', y='Fare', data=data)
plt.title('Survival by Fare')
plt.show()
```

```python
# Embarked distribution
sns.countplot(x='Embarked', data=data)
plt.title('Port of Embarkation Distribution')
plt.show()
```

```python
# Relationship between embarked and survival
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title('Survival by Port of Embarkation')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
